```python
# Ejercicio basado en entrenamiento de imagenes a color de la base de datos 
# CIFAR 10 que contiene  imagenes de 10 diferentes tipos, dentro de las cuales se encuentran 
# imagenes de entrenamiento e imagenes de prueba.
# el codigo siguiente muestra el entrenamiento utilizando Tensorflow con Keras

# PARTE 1 

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalize pixel values to be between 0 and 1

train_images=train_images.reshape(50000, 32, 32, 3)
train_images=train_images / 255.0
test_images = test_images.reshape(10000, 32, 32, 3)
test_images=test_images/255.0

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Modificando  el tamao de las imgenes al momento de visualizarlas entre 2 y 2.5  veces
xSize = 20;
ySize = 20;
fontSize = 18;
numberOfImages = 14;

# Mtodo que nos permite clasificar imgenes con etiquetas en espaol

# You could use a dictionary:
def getTagName(x):
    return {
        0: 'avion',
        1: 'automovil',
        2: 'ave',
        3: 'gato',
        4: 'venado',
        5: 'perro',
        6: 'rana',
        7: 'caballo',
        8: 'barco',
        9: 'camin',
    }[x]

def getArrayName(x):
    return {
        'airplane': 'avion',
        'automobile': 'automovil',
        'bird': 'ave',
        'cat': 'gato',
        'deer': 'venado',
        'dog': 'perro',
        'frog': 'rana',
        'horse': 'caballo',
        'ship': 'barco',
        'truck': 'camin',
    }[x]


plt.figure(figsize=(xSize - 5, ySize - 5))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    # The CIFAR labels happen to be arrays, 
    # which is why you need the extra index
    plt.xlabel(getTagName(train_labels[i][0]), fontsize=fontSize)
plt.show()

# Aqui se requiere incluir las convoluciones

model = models.Sequential()
model.add(layers.Conv2D(16, (3, 3), activation='sigmoid', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(16, (3, 3), activation='sigmoid'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(16, (3, 3), activation='sigmoid'))
model.add(layers.Flatten())
model.add(layers.Dense(20, activation='sigmoid'))
model.add(layers.Dense(10, activation='softmax'))

model.summary()

model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=2, 
                    validation_data=(test_images, test_labels))

plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')

test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)

print(test_acc)

model.save('/tmp/model7')

#Parte 2 -Prueba con 15 imagenes aleatorias (incluir en otro cuadro de codigo despues de correr el primero)

import numpy as np
from random import randint, uniform,random

num_1 = randint(0,9989)
num_2 = num_1 + numberOfImages

arr = np.arange(num_1,num_2).reshape(numberOfImages)
permutation = np.random.permutation(arr)
test_images1, test_labels1 = (test_images[permutation], test_labels[permutation])

predictions = model.predict(test_images1)
evaluations = model.evaluate(test_images1, test_labels1)

arreglo = (np.array(class_names)[np.argmax(predictions, axis=-1)].tolist())
color = 'red'

plt.figure(figsize=(xSize,ySize))
for i in range(numberOfImages):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(test_images1[i], cmap=plt.cm.binary)
     
    if  (class_names[test_labels1[i][0]]).lower() == arreglo[i].lower():
       color ='green'
    else:
       color ='red'

    plt.xlabel(getTagName(train_labels[i][0])+" (p="+ getArrayName(arreglo[i]) +")", fontsize=fontSize).set_color(color)
    
plt.show()
```
